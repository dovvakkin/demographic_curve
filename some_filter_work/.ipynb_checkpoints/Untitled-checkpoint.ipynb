```python
def good_str(line):
    output = line[5]
    #print(line[5])
    for i in range(6, len(line)):
        output += ' {}'.format(line[i])
    return output
```

```python
f = open("data_fullnesses.txt", 'r')
csvs_list = []
for line in f:
    str = line.split()
    #print(str[3], str[5])
    if len(str) > 5:
        if float(str[3]) > 0.65:
            csvs_list.append(good_str(str))
        
```
